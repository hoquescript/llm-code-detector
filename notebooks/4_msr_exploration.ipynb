```python
import numpy as np
import pandas as pd
import pyarrow.parquet as pq
```

#### Reading Data

```python
# All datasets
all_pull_request = pd.read_parquet("../data/msr/all_pull_request.parquet")
all_repository = pd.read_parquet("../data/msr/all_repository.parquet")
all_user = pd.read_parquet("../data/msr/all_user.parquet")

# Main datasets
pull_request = pd.read_parquet("../data/msr/pull_request.parquet")
repository = pd.read_parquet("../data/msr/repository.parquet")
user = pd.read_parquet("../data/msr/user.parquet")
issue = pd.read_parquet("../data/msr/issue.parquet")

# PR-related datasets
pr_comments = pd.read_parquet("../data/msr/pr_comments.parquet")
pr_commits = pd.read_parquet("../data/msr/pr_commits.parquet")
pr_commit_details = pd.read_parquet("../data/msr/pr_commit_details.parquet")
pr_reviews = pd.read_parquet("../data/msr/pr_reviews.parquet")
pr_review_comments = pd.read_parquet("../data/msr/pr_review_comments.parquet")
pr_review_comments_v2 = pd.read_parquet("../data/msr/pr_review_comments_v2.parquet")
pr_task_type = pd.read_parquet("../data/msr/pr_task_type.parquet")
pr_timeline = pd.read_parquet("../data/msr/pr_timeline.parquet")

# Human datasets
human_pull_request = pd.read_parquet("../data/msr/human_pull_request.parquet")
human_pr_task_type = pd.read_parquet("../data/msr/human_pr_task_type.parquet")

# Related datasets
related_issue = pd.read_parquet("../data/msr/related_issue.parquet")
```

##### EDA (Exploratory Data Analysis)

```python
datasets = {
    "all_pull_request": all_pull_request,    
    "all_repository": all_repository,
    "all_user": all_user,
    "pull_request": pull_request,
    "repository": repository,
    "user": user,
    "issue": issue,
    "pr_comments": pr_comments,
    "pr_commits": pr_commits,
    "pr_commit_details": pr_commit_details,
    "pr_reviews": pr_reviews,
    "pr_review_comments": pr_review_comments,
    "pr_review_comments_v2": pr_review_comments_v2,
    "pr_task_type": pr_task_type,
    "pr_timeline": pr_timeline,
    "human_pull_request": human_pull_request,
    "human_pr_task_type": human_pr_task_type,
    "related_issue": related_issue
}

print("Dataset Overview:")
print("=" * 80)
for name, df in datasets.items():
    print(f"{name:30s} | Rows: {len(df):>10,} | Columns: {len(df.columns):>3} | Memory: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
```

```python
print("Column Overview:")
print("=" * 80)
for name, df in datasets.items():
    print(f"{name}({len(df.columns)}) | {df.shape}")
    print(f"{df.columns.tolist()}")
    print("-" * 80)
```
